5. I then learned that web scraping with dynamic content (like clicking links) often requires a tool like 'Selenium', which can automate interactions with a webpage. 

6. I then installed 'Selenium' to my editor

7. I also download the Chrome WebDriver for 'Selenium' according to my Chrome versio which at the time was 117.0.5938.150.
